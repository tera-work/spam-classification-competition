```python
import pandas as pd
import numpy as np
import glob

train = pd.read_csv('train_master.csv', index_col=0)
train.head()
train.info()
train.describe()
```

```python
len(glob.glob('train2/train_*.txt'))
len(glob.glob('test2/test_*.txt'))
```

```python
train['text_length'] = train['text'].apply(len)
train['num_words'] = train['text'].apply(lambda x: len(x.split()))
train[['text_length','num_words']].describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x=train['label'])
plt.show()
```

```python
from collections import defaultdict

corpus = []
for x in train['text'].str.split():
    for i in x:
        corpus.append(i)

dic = defaultdict(int)
for word in corpus:
    dic[word] += 1

top = sorted(dic.items(), key=lambda x:x[1], reverse=True)[:15]
x, y = zip(*top)
plt.bar(x,y)
plt.show()
```
